**Increasing the number of hidden layers**

```python
mse_normalized=[]

for i in range(50):

  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)

  n_cols=X_train.shape[1]
  model=Sequential()
  model.add(Dense(10,activation="relu",input_shape=(n_cols,)))
  model.add(Dense(10,activation="relu"))
  model.add(Dense(10,activation="relu"))
  model.add(Dense(1))
  model.compile(optimizer="adam",loss="mean_squared_error")
  model.fit(X_train,y_train,epochs=50,verbose=0)
  y_predict=model.predict(X_test)

  mse_error=mean_squared_error(y_test,y_predict)
  mse_normalized.append(mse_error)
```

```python
hmean_normalized=np.mean(mse_error)
hsd_normalized=np.std(mse_error)
print("normal mean is {}".format(mean))
print("normal std is {}".format(sd))
print("normalized mean is {}".format(mean_normalized))
print("normalized std is {}".format(sd_normalized))
print("normalized mean after increasing epoch is {}".format(emean_normalized))
print("normalized std after increasing epoch is {}".format(esd_normalized))
print("normalized mean after increasing hidden layer is {}".format(hmean_normalized))
print("normalized std after increasing hidden layeris {}".format(hsd_normalized))
```
